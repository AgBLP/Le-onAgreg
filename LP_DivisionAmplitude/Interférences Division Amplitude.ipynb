# Interférences à division d'amplitude

## 1) Intensité à travers une lame d'air

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Intensite_lame(e,n,lamb,i):
    return(0.5*(1+np.cos(4*np.pi*n*e*np.cos(i)/lamb)))
```

```python
#Paramètres initiaux
e = 0.2e-3
n=1 #air
lamb = 638e-9
```

```python
i = np.linspace(-10*np.pi/180,10*np.pi/180,5001)
I = Intensite_lame(e,n,lamb,i)
```

```python
#Affichage résultats
fig1 = plt.plot(i*180/np.pi,I,label = r'e=0.2mm, $\lambda=638$nm')
plt.title("Intensité en fonction de l'angle d'incidence sur la lame d'air pour une source monochromatique",fontsize=15,pad=20, fontweight= 'bold')
plt.rc('font', size=10)
plt.rc('axes', labelsize=15)
plt.xlabel(r'i (en $\degree$)')
plt.ylabel(r'I/I$_0$')
plt.legend(bbox_to_anchor=(1, 0, 0, 1))
```

## 2) Interférences pour une source duo-chromatique

```python
def interf_duo(e,n,lamb1,lamb2,i):
    mean_lamb = 0.5*(lamb1+lamb2)
    diff_lamb = abs(lamb1-lamb2)
    delta = 2*n*e*np.cos(i)
    return(0.5*(1+np.cos(2*np.pi*delta/mean_lamb)*np.cos(np.pi*delta*diff_lamb/(mean_lamb*mean_lamb))))
```

```python
#Paramètres initiaux
i = 1*np.pi/180
n=1 #air
lamb1 = 589e-9
lamb2 = 589.6e-9
```

```python
e = np.linspace(-1.9e-4,1.9e-4,5001)
I = interf_duo(e,n,lamb1,lamb2,i)
```

```python
#Affichage résultats
fig1 = plt.plot(e*1000,I, label= r'i=1$\degree$, $\lambda_{1}=589$nm, $\lambda_{2}=589.6$nm')
plt.title("Intensité en fonction de l'épaisseur de la lame d'air pour une source duochromatique", fontsize=15,pad=20, fontweight= 'bold')
plt.rc('font', size=10)
plt.rc('axes', labelsize=15)
plt.xlabel('e (en mm)')
plt.ylabel(r'I/I$_0$')
plt.legend(bbox_to_anchor=(1, 0, 0, 1), ncol=3,borderpad = 0.5,labelspacing = 0.5)
```
